```python
from ultralytics import YOLO  # YOLO 라이브러리 임포트
import cv2 as cv

# YOLO 모델 로드
model = YOLO('yolo11n.pt')

# 이미지 읽기
img = cv.imread("images/HANDSOME.jpg")

# YOLO 모델로 예측
results = model.predict(img, conf=0.5)

# 가장 큰 박스 찾기
largest_area = 0
largest_box_idx = -1

for result in results:
    for idx, box in enumerate(result.boxes):
        # 박스 정보
        x1, y1, x2, y2 = box.xyxy[0]  # 좌표 (x1, y1), (x2, y2)
        cls_id = int(box.cls)  # 클래스 ID

        # "person" 클래스에 대해 처리
        if result.names[cls_id] == "person":
            # 박스 크기 계산 (w * h)
            area = (x2 - x1) * (y2 - y1)
            if area > largest_area:
                largest_area = area
                largest_box_idx = idx

# 결과 시각화 및 라벨 변경
for result in results:
    for idx, box in enumerate(result.boxes):
        # 박스 정보
        x1, y1, x2, y2 = box.xyxy[0]
        cls_id = int(box.cls)
        conf = float(box.conf)  # confidence 값

        # "person" 클래스의 라벨 처리
        if result.names[cls_id] == "person":
            if idx == largest_box_idx:
                label = "Seo Kang Jun (0.99)"  # 가장 큰 박스
            else:
                label = f"person ({conf:.2f})"  # 나머지 박스
            
            # 박스와 라벨 그리기
            x1, y1, x2, y2 = map(int, [x1, y1, x2, y2])  # 좌표 정수 변환
            cv.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)  # 박스 그리기
            cv.putText(img, label, (x1, y1 - 10), cv.FONT_HERSHEY_SIMPLEX, 
                       0.5, (0, 255, 0), 2)  # 라벨 텍스트 추가

# 결과 표시
cv.imshow("img", img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
from ultralytics import YOLO  # YOLO 라이브러리 임포트
import cv2 as cv

cap = cv.VideoCapture("images/cars.mp4")
fps = cap.get(cv.CAP_PROP_FPS)
   
while cap.isOpened():  
    ret, frame = cap.read()
    if not ret:
        break
    result = model.predict(frame, conf=0.4)
    frame = result[0].plot()    

   
    cv.imshow("Video", frame)

    if cv.waitKey(int(1000 / fps)) == ord("q"):
        break

cap.release()
cv.destroyAllWindows()
```
