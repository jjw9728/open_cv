## 3.1 간단하게

```python
import cv2 as cv

img = cv.imread("images/BB.jpg", cv.IMREAD_GRAYSCALE)

ret, binary = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
ret, otsu = cv.threshold(img, -1, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)

cv.imshow("img", img)
cv.imshow("binary", binary)
cv.imshow("otsu", otsu)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2 as cv

# 전역 변수 설정
img = cv.imread("images/BB.jpg")
points = [] 

def mouse_handler(event, x, y, flags, param):
    global points
    if event == cv.EVENT_LBUTTONDOWN:  
        points.append((x, y))  
        cv.circle(img, (x, y), 3, (0, 0, 255), cv.FILLED) 
        if len(points) > 1:  
            cv.line(img, points[-2], points[-1], (255, 0, 0), 2)
        cv.imshow("img", img)

cv.namedWindow("img")
cv.setMouseCallback("img", mouse_handler)

cv.imshow("img", img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np

img = cv.imread("images/card.jpg")
point_list = [] 

def onMouse(event, x, y, flags, param):
    global point_list
    if event == cv.EVENT_LBUTTONDOWN: 
        if len(point_list) < 4: 
            point_list.append((x, y))
        if len(point_list) == 4: 
            show_result()

def show_result():
    global point_list
    if len(point_list) == 4:        
        src = np.array(point_list, dtype=np.float32)
        width, height = 400, 300 
        dst = np.array([[0, 0],
                        [width - 1, 0],
                        [width - 1, height - 1], 
                        [0, height - 1]],
                        dtype=np.float32)
        
        matrix = cv.getPerspectiveTransform(src, dst)
        scanned = cv.warpPerspective(img, matrix, (width, height))
        
        cv.imshow("Scan", scanned)

name = "Scanner"
cv.namedWindow(name)
cv.setMouseCallback(name, onMouse)

while True:
    scan_img = img.copy() 
    for i, point in enumerate(point_list):  
        cv.circle(scan_img, point, 5, (255, 0, 0), -1)
        if i > 0:
            cv.line(scan_img, point_list[i - 1], point, (255, 0, 0), 2)    
  
    cv.imshow(name, scan_img)
    if cv.waitKey(1) == ord("q"):
        break

cv.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np

img = cv.imread("images/card.jpg")
img = cv.resize(img, (900, 700))
point_list = []

def show_result():
    global point_list
    if len(point_list) == 4:
        src = np.array(point_list, dtype=np.float32)
        width, height = 400, 300
        dst = np.array([[0, 0],
                        [width, 0],
                        [width, height],
                        [0, height]], dtype=np.float32)
        
        matrix = cv.getPerspectiveTransform(src, dst)
        scanned = cv.warpPerspective(img, matrix, (width, height))
        cv.imshow("Scan", scanned)

def onMouse(event, x, y, flags, _):
    global point_list, mouse_x, mouse_y
    if event == cv.EVENT_LBUTTONDOWN:
        if len(point_list) < 4:
            point_list.append((x, y))
        if len(point_list) == 4:
            show_result()
    elif event == cv.EVENT_MOUSEMOVE:
        mouse_x, mouse_y = x, y

name = "Scanner"
cv.namedWindow(name)
cv.setMouseCallback(name, onMouse)

mouse_x, mouse_y = 0, 0

while True:
    scanned_img = img.copy()
    for i, point in enumerate(point_list):
        cv.circle(scanned_img, point, 5, (255, 0, 0), cv.FILLED)
        if i > 0:
            cv.line(scanned_img, point_list[i - 1], point, (255, 0, 0), 2)
    if len(point_list) > 0 and len(point_list) < 4:
        cv.line(scanned_img, point_list[-1], (mouse_x, mouse_y), (0, 255, 0), 2)
    if len(point_list) == 4:
        cv.line(scanned_img, point_list[-1], point_list[0], (255, 0, 0), 2)
    cv.imshow(name, scanned_img)
    if cv.waitKey(1) != -1:
        break

cv.imshow(name, img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread("images/card.jpg")
point_list = []

# 최종 결과를 출력하는 함수입니다
def show_result():
  width, height = 600, 350
  src = np.array(point_list, dtype=np.float32)
  # src = np.float32(point_list)
  target = np.array([[0,0], [width, 0], [width, height], [0, height]], dtype=np.float32)
  mat = cv2.getPerspectiveTransform(src, target)
  result = cv2.warpPerspective(img, mat, (width, height))
  cv2.imshow("result", result)

# 마우스 이벤트와 함께 실행되는 콜백함수입니다.
drawing = False
def on_mouse(event, x, y, flags, _):
  global drawing
  coppied = img.copy()
  if event == cv2.EVENT_LBUTTONDOWN:
    drawing = True
    point_list.append((x,y))

  if drawing:
    prev_point = None
    for point in point_list:
      cv2.circle(coppied, point, 10, (0,0,255), cv2.FILLED)
      if prev_point:
        cv2.line(coppied, prev_point, point, (0,0,255), 5)
      prev_point = point

    next_point = (x,y)
    if len(point_list) == 4:
      next_point = point_list[0]
      show_result()
    cv2.line(coppied, point_list[-1], next_point, (0,0,255), 5)

  cv2.imshow(name, coppied)

# 기본적인 코드 실행입니다.
name = "Scanner"
cv2.namedWindow(name)
cv2.setMouseCallback(name, on_mouse)
cv2.imshow(name, img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
