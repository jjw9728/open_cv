# 10. 히스토그램


```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("images/BB.jpg")
col = ["b", "g", "r"]

for i in range(len(col)):
    hist = cv.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=col[i])
    
plt.title("histogram")
plt.xlabel("Pixel value")
plt.ylabel("Frequency")
plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("images/BB.jpg")
color = ('b', 'g', 'r') 

for i, col in enumerate(color):
    hist = cv.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=col)

plt.title("RGB Histogram")
plt.xlabel("Pixel value")
plt.ylabel("Frequency")
plt.show()
```

# 11. 경계선 검출

### 기본 구현


```python
img = cv.imread("images/BB.JPG")
canny = cv.Canny(img, 50, 150)

cv.imshow("img", img)
cv.imshow("canny", canny)
cv.waitKey(0)
cv.destroyAllWindows()
```

### 트랙바 사용

```python
import cv2 as cv

def empty(_):
    pass

img = cv.imread("images/mountain.jpg", cv.IMREAD_GRAYSCALE)

name = "Trackbar"
cv.namedWindow(name)
cv.createTrackbar("threshold1", name, 0, 255, empty)
cv.createTrackbar("threshold2", name, 0, 255, empty)

while True:
    threshold1 = cv.getTrackbarPos("threshold1", name)
    threshold2 = cv.getTrackbarPos("threshold2", name)
    canny = cv.Canny(img, threshold1, threshold2)
    cv.imshow("canny", canny)

    if cv.waitKey(1) != -1:
        break

cv.destroyAllWindows()
```

### 웹캠 캐니

```python
import cv2 as cv

cap = cv.VideoCapture(0)
if not cap.isOpened():
    exit()

name = "Canny"
cv.namedWindow(name)
cv.createTrackbar("threshold1", name, 50, 255, lambda x:x)
cv.createTrackbar("threshold2", name, 150, 255, lambda x:x)

while True:  
    ret, frame = cap.read()
    
    if not ret:        
        break 
      
    threshold1 = cv.getTrackbarPos("threshold1", name)
    threshold2 = cv.getTrackbarPos("threshold2", name)
    canny = cv.Canny(frame, threshold1, threshold2)  
    cv.imshow(name, canny)

    if cv.waitKey(1) != -1:
        break

cap.release()
cv.destroyAllWindows()
```
